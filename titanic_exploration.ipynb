## Data Cleaning

#### After exploring the dataset, we can now begin to clean it. I am creating a new dataset to work on, which will afford me the freedom of mistakes.

#### Our new datset is called 'df_new'

#### I am creating a copy of the dataset using .copy() rather than placing the dataset in a variable. Placing the dataset in a new variable means that any changes made will still affect the dataset.

```python
# Creating a direct copy of the dataset
df_new = df.copy()
df_new
```

Here I am dropping the column 'deck' as it contains too many missing values and it does not seem to be of much value as a column

```python
# Using 'inplace=True' to directly alter the dataset
df_new.drop(columns=['deck'], inplace=True)
```

```python
df_new.head(50)
```

```python
df_new.shape
```

```python
# Getting a new view of the amount of missing values in each row
missing_data = df_new.isna().sum()
print(missing_data)
```

Returning all the rows in the dataset & whether the value in column 'age' is missing or not and placing them in a variable to use

```python
missing_ages = df_new['age'].isna()
# print(missing_ages) # Just checking
```

Trying to view only the rows where the 'age' value is missing
using the boolean value- returning the rows where '.isna()' returns as 'True'

```python
missing_ages_rows = df_new[missing_ages]
# print(missing_ages_rows) - prints a not-so-pretty view of all the 'missing_ages_rows'
missing_ages_rows.head(60)
```

```python
# Dropping those two rows & getting updated row/coloumn numbers
df_new.dropna(subset=['embark_town'], inplace=True)
df_new.shape
```

In this section, I will be replacing values in the dataset.

```python
# Replacing '1' and '0' with Yes and No
df_new['survived'] = df_new['survived'].replace(1, "Yes")
df_new['survived'] = df_new['survived'].replace(0, "No")
df_new
# (When using inplace, returns 'None"- why?)
```

```python
# Replacing values '1, 2, 3' with their corresponding class titles
df_new['pclass'] = df_new['pclass'].replace(1, "Upper")
df_new['pclass'] = df_new['pclass'].replace(2, "Middle")
df_new['pclass'] = df_new['pclass'].replace(3, "Lower")
df_new.head(10)
```

In this section, I will be reformatting the numerical data values.

```python
# Casting values for 'age' as int to remove decimals
df_new['age']= df_new['age'].astype(int)
df_new.head(10)
```

```python
# Removing the extra decimals from the 'fare' values
df_new['fare'] = df_new['fare'].round()
df_new.head(10)
```

Now that we're done with the main cleaning, I'll be renaming the columns.

```python
# Renaming the neccessary columns
df_new.rename(columns={'pclass': 'social_class', 'embarked': 'embarked_code', 'class': 'ticket_class', 'who': 'demographic'}, inplace=True)
df_new.head(50)
```
